1. Procedural Programming

focuses on procedures or functions. You write a list of instructions for the computer to follow. The data is usually stored in variables and passed around from one function to another.

Focus: Logic and flow.

Structure: Top-down.

it is like a "To-Do" list. If you want to wash a car, you write: soak(), scrub(), rinse(), dry().

2. Object-Oriented Programming (OOP)

OOP organizes code around data, or objects, rather than functions and logic. An object is a self-contained unit that holds both data (attributes) and behaviors (methods).

Focus: Data objects and how they interact.

Structure: Bottom-up.

It is like a Car object. It has attributes like color and fuel_level, and it knows how to drive() or honk().

```python
class Employee:
    def __init__(self,age, name):
        print("my first class")
        self.age = age
        self.name = name
    def pname(self):
        print("my name is " + self.name)
    def page(self):
        print("my age is " + str(self.age))
```

```python
emp_1 = Employee(30, "Alice")
```

```python
class person:
    def __init__(self, name, age, height, weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight
    def pname(self):
        print("my name is " + self.name)
    def page(self):
        print("my age is " + str(self.age))
    def pheight(self):
        print("my height is " + str(self.height))
    def pweight(self):
        print("my weight is " + str(self.weight))

p1 = person("John", 25, 175, 70)
p2 = person("Yusuf", 30, 165, 60)
p3 = person("Ahmed", 28, 180, 80)

p1.pname()
p2.page()
p3.pheight()
p1.pweight()

```
