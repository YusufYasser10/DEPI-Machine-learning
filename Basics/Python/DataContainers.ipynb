```python
#Lists are ordered, mutable, and allow duplicate elements, allow heterogeneous data types.
fruits = ["apple", "banana", "cherry", "date", 10, 15.5, True]
```

```python
#list operations
# indexing
print(fruits[0])        
print(fruits[-1])      
# slicing
print(fruits[1:4])      
# appending
fruits.append("elderberry")
print(fruits)
# removing
fruits.remove("banana")
print(fruits)
```

```python
#tuples are ordered, immutable, and allow duplicate elements, allow heterogeneous data types.
coordinates = (10, 20, 30, "A", "B", "C")
#tuple operations
# indexing
print(coordinates[1])    # Output: 20
# slicing
print(coordinates[2:5])  # Output: (30, 'A', 'B')
# counting occurrences
print(coordinates.count("A"))  # Output: 1
# finding index
print(coordinates.index(30))    # Output: 2
```

```python
#sets are unordered, mutable, and do not allow duplicate elements, allow heterogeneous data types.
unique_numbers = {1, 2, 3, 4, 5, 3, 2, 1}
# set operations
# adding elements
unique_numbers.add(6)
print(unique_numbers)        # Output: {1, 2, 3, 4, 5, 6}
```

```python
# 1. Non-parameterized function without return
def greet():
    print("Hello, World!")

greet()  # Output: Hello, World!
```

```python
# 2. Non-parameterized function with return
def get_pi():
    return 3.14159
result = get_pi()
print(f"Pi value: {result}")

#return statements = keywords or instructions send python values/objects back to the caller.
```

```python
# 3. Parameterized function without return
def greet_person(name, age):
    print(f"Hello, {name}! You are {age} years old.")

greet_person("Alice", 25)  # Output: Hello, Alice! You are 25 years old.
greet_person("Bob", 30)    # Output: Hello, Bob! You are 30 years old.
```

```python
# 4. Parameterized function with return
def add_numbers(a, b):
    return a + b

sum_result = add_numbers(10, 20)
print(f"Sum: {sum_result}")  # Output: Sum: 30

def calculate_area(length, width):
    area = length * width
    return area

rectangle_area = calculate_area(5, 8)
print(f"Area: {rectangle_area}")  # Output: Area: 40
```

```python
# Summary comparison
print("=" * 50)
print("FUNCTION TYPES COMPARISON")
print("=" * 50)

# Non-parameterized without return - just performs an action
def type1():
    print("Type 1: No parameters, no return")
type1()

# Non-parameterized with return - returns a fixed value
def type2():
    return "Type 2: No parameters, has return"
print(type2())

# Parameterized without return - performs action based on input
def type3(x):
    print(f"Type 3: Has parameters ({x}), no return")
type3(100)

# Parameterized with return - processes input and returns result
def type4(x, y):
    return f"Type 4: Has parameters ({x}, {y}), has return: {x * y}"
print(type4(5, 7))
```

```python
def find_factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Usage:
print(find_factorial(5)) # Output: 120
```

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5)) 
```

```python
def fibonacci(n):
    if n <= 0:
        return "Please enter a positive integer"
    
    a, b = 0, 1
    series = []

    for _ in range(n):
        series.append(a)
        a, b = b, a + b

    return series
print(fibonacci(10))
```
