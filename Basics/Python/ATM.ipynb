1- Procedural programming

Customer Requirements for Traditional Bank Account Management System
Single Account: The system should manage only one account (in this case, Alice's account). The account starts with an initial balance, which can be modified.
Basic Operations: The customer should be able to: Deposit money into the account. Withdraw money from the account. Check the current balance.
Balance Monitoring: The system should allow the customer to see their balance after each deposit or withdrawal.
Simple Error Handling: If the customer tries to withdraw more than the balance, the system should show an "Insufficient funds" message.
Simple Interface: The system should have basic functions for depositing, withdrawing, and checking the balance, with easy interaction.
Example:
Alice starts with a balance of 1000.
She deposits 500, and her balance becomes 1500.
She withdraws 200, and her balance becomes 1300.
She can check the balance, which shows 1300.

```python
balance = 1000
pin = "1234"

def check_balance():
    print(f"Your current balance is: ${balance}")

def deposit(amount):
    global balance
    balance += amount
    print(f"${amount} deposited successfully, new balance is: ${balance}")

def withdraw(amount):
    global balance
    if amount > balance:
        print("Insufficient funds!")
    else:
        balance -= amount
        print(f"${amount} withdrawn successfully, new balance is: ${balance}")

user_pin = input("Enter PIN: ")
if user_pin == pin:
    check_balance()
    deposit(500)
    withdraw(200)
    check_balance()
else:
    print("Invalid PIN")
```

2- Customer's New Requirements

The customer has now asked for the following changes:
Multiple accounts: The system should allow multiple customers, each with their own balance and account information.
Account types: There should be different account types like Savings Account and Joint Account.
Interest calculation: Savings accounts need to apply an interest rate.
Error handling: The system should be able to handle different types of errors in a cleaner way.
Scalability: As the bank grows, the system should scale easily without needing a complete rewrite.

```python
class BankAccount:
    def __init__(self, owner, account_type, balance=0):
        self.owner = owner
        self.account_type = account_type
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"[{self.owner}] +${amount}. New Balance: ${self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"[{self.owner}] Error: Insufficient funds!")
        else:
            self.balance -= amount
            print(f"[{self.owner}] -${amount}. New Balance: ${self.balance}")

    def apply_interest(self):
        if self.account_type == "Savings":
            interest = self.balance * 0.02
            self.deposit(interest)
            print(f"Interest of ${interest} applied.")
        else:
            print("Interest only applies to Savings Accounts.")


# --- Using the system ---
Ahmed = BankAccount("Ahmed", "joint", 1000)
Yusuf = BankAccount("Yusuf", "Savings", 2000)
Yusuf.withdraw(300)
Ahmed.deposit(400)
Yusuf.apply_interest()   
Ahmed.apply_interest()
```
