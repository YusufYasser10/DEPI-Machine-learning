```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5)) 
```

```python
def fibonacci(n):
    if n <= 0:
        return "Please enter a positive integer"
    
    a, b = 0, 1
    series = []

    for _ in range(n):
        series.append(a)
        a, b = b, a + b

    return series
print(fibonacci(10))
```

```python
def average(grades):
    if grades == []:
        return 0
    total = 0
    for g in grades:
        total += g
    return total / len(grades)

def get_status(avg):
    if avg >= 85:
        return "Excellent"
    elif avg >= 70:
        return "Good"
    elif avg >= 50:
        return "Pass"
    else:
        return "Fail"

def process_students(data):
    student_results = {}
    for name, grades in data:
        avg = average(grades)
        status = get_status(avg)
        student_results[name] = status , avg
    return student_results

data = [
    ("Alice", [90, 85, 88]),
    ("Bob", [70, 65, 72]),
    ("Charlie", [50, 45, 52]),
    ("Daisy", [30, 40, 25])
]

output = process_students(data)
print(output)
```

```python
products_list = [("Laptop", 1200), ("Mouse", 25), ("Keyboard", 50), ("Monitor", 200)]
products_dict = {name: price for name, price in products_list}

def show_products(products_dict):
    for product, price in products_dict.items():
        print(f"{product}: ${price}")

def add_to_cart(cart, product_name, quantity, products_dict):
    if product_name in products_dict:
        if product_name in cart:
            cart[product_name] += quantity
            print(f"Updated {product_name} quantity to {cart[product_name]}.")
        else:
            cart[product_name] = quantity
            print(f"Added {quantity}x {product_name} to cart.")
    else:
        print(f"Error: '{product_name}' is not available in our store.")

def calculate_total(cart, products_dict):
    total = 0
    for product, quantity in cart.items():
        price = products_dict[product]
        total += price * quantity
    return total

user_cart = {}
show_products(products_dict)
add_to_cart(user_cart, "Laptop", 1, products_dict)
add_to_cart(user_cart, "Mouse", 2, products_dict)
add_to_cart(user_cart, "Mouse", 1, products_dict) 
grand_total = calculate_total(user_cart, products_dict)
print(f"Final Cart: {user_cart}")
print(f"Total Amount Due: ${grand_total}")
```

```python
real_password = "1234"
attempts = 0
status = False
print("--- Welcome to the ATM ---")

while attempts < 3 and status == False:
    user_input = input(f"Enter password (Attempt {attempts + 1}/3): ")
    status = (user_input == real_password)
    message = ["Wrong password, please enter the correct password", "Success"][status]
    print(message)
    attempts += 1
is_blocked = (status == False) and (attempts == 3)
blocked_msg = ["", "Your card is blocked! Please contact the bank."][is_blocked]
print(blocked_msg)
```
