```python
import pandas as pd

employees = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Age': [28, 34, 29, 42, 36, 50, 45],
    'Address': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio'],
    'Salary': [70000, 80000, 65000, 90000, 100000, 410000, 300000]
}

data_frame = pd.DataFrame(employees)
print("Original DataFrame:")
print(data_frame)


print("\n\nSelected rows (0-2) with specific columns:")
print(data_frame.loc[0:2, ['Name', 'Age', 'Address']])


data_frame['Major'] = ['Engineer', 'Doctor', 'Pharmacist', 'Engineer', 'Lawyer', 'Artist', 'Architect']
print("\n\nDataFrame after adding 'Major' column:")
print(data_frame)


data_frame['Salary'] = 10000
print("\n\nDataFrame after updating Salary to 10000:")
print(data_frame)


data_frame['Add'] = data_frame['Address']
print("\n\nDataFrame after duplicating Address to Add:")
print(data_frame)


data_frame = data_frame.drop('Address', axis=1)
print("\n\nFinal DataFrame after dropping 'Address' column:")
print(data_frame)
```
