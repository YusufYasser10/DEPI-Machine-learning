```python
def plot_num_dist(df, cols):
    for col in cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            plt.figure(figsize=(8,4))
            sns.histplot(df[col].dropna(), bins=60)
            plt.title(f'Distribution of {col}')
            plt.xlabel(col)
            plt.ylabel('Frequency')
            plt.show()

num_cols = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits', 'TotalCompensation']
plot_num_dist(salaries, num_cols)
```

```python
# looking through the data to understand which is a majority: Full-time or part-time, which job title is the most frequent and so on
def plot_cat_counts(df, cols, top_n=10):
    for col in cols:
        if col in df.columns:
            order = df[col].value_counts().index[:top_n]
            sns.countplot(data=df, x=col, order=order)
            plt.title(f'Top {col} Counts')
            plt.xticks(rotation=45)
            plt.show()

cat_cols = ['JobTitle', 'Year', 'Status']
plot_cat_counts(salaries, cat_cols)
```

```python
# Correlation matrix
corr = salaries.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# looking at IQR and its values as box plot
def plot_outliers(df, cols):
    for col in cols:
        if col in df.columns:
            plt.figure(figsize=(8,4))
            sns.boxplot(x=df[col].dropna())
            plt.title(f'Boxplot of {col}')
            plt.show()
            # IQR method
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
            print(f'{col}: {len(outliers)} outliers')

salary_cols = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits', 'TotalCompensation']
plot_outliers(salaries, salary_cols)
```

```python
# which job has the most Total Pay benefits
if 'TotalPayBenefits' in salaries.columns:
    display(salaries.nlargest(10, 'TotalPayBenefits'))

# Average salary for a JobTitle
if 'JobTitle' in salaries.columns and 'TotalPayBenefits' in salaries.columns:
    avg_by_job = salaries.groupby('JobTitle')['TotalPayBenefits'].mean().sort_values(ascending=False).head(10)
    plt.figure(figsize=(12,6))
    sns.barplot(x=avg_by_job.values, y=avg_by_job.index)
    plt.title('Top 10 Job Titles by Average TotalPayBenefits')
    plt.xlabel('Average TotalPayBenefits')
    plt.ylabel('Job Title')
    plt.show()
```

## Business Relation
over-time vs another full-time shift, If the bussiness owner wants to cancel overtime, how many employees he needs to hire (approximation)

```python
top_overtime = salaries.groupby('JobTitle')['OvertimePay'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_overtime.values, y=top_overtime.index, palette='magma')
plt.title('Top 10 Job Titles by Average Overtime Pay')
plt.xlabel('Average Overtime Pay')
plt.ylabel('Job Title')
plt.show()

high_overtime = salaries[salaries['OvertimePay'] > salaries['BasePay']]
print(f'Number of employees with overtime pay exceeding base pay: {len(high_overtime)}')

total_overtime = salaries['OvertimePay'].sum()
avg_base = salaries['BasePay'][salaries['BasePay'] > 0].mean()
potential_new_hires = int(total_overtime // avg_base)
print(f'Total overtime paid: ${total_overtime:,.2f}')
print(f'Average base salary: ${avg_base:,.2f}')
print(f'Potential number of new hires (if all overtime was replaced by new staff): {potential_new_hires}')
```

```python
# how we can handle missing data

salaries['BasePay'] = pd.to_numeric(salaries['BasePay'], errors='coerce').fillna(0)  
salaries['OvertimePay'] = pd.to_numeric(salaries['OvertimePay'], errors='coerce').fillna(0)  
salaries['OtherPay'] = pd.to_numeric(salaries['OtherPay'], errors='coerce').fillna(0) 
salaries['Benefits'] = pd.to_numeric(salaries['Benefits'], errors='coerce').fillna(0) 


salaries['TotalPay'] = salaries['BasePay'] + salaries['OvertimePay'] + salaries['OtherPay']
salaries['TotalPayBenefits'] = salaries['TotalPay'] + salaries['Benefits']


salaries['JobTitle'] = salaries['JobTitle'].fillna('Unknown')
salaries['Status'] = salaries['Status'].fillna('Unknown')
salaries['Notes'] = salaries['Notes'].fillna('Not Provided')
salaries['EmployeeName'] = salaries['EmployeeName'].fillna('Unknown')


if salaries['Agency'].isnull().any():
    mode_agency = salaries['Agency'].mode()[0]
    salaries['Agency'] = salaries['Agency'].fillna(mode_agency)


salaries = salaries.dropna(subset=['Year'])


print('Remaining missing values after imputation:')
print(salaries.isnull().sum())
```
